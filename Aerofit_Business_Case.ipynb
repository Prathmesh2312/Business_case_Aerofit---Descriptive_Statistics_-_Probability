Aerofit is a leading brand in the field of fitness equipment. Aerofit provides a product range including machines such as treadmills, exercise bikes, gym equipment, and fitness accessories to cater to the needs of all categories of people.

### ‚ö†Ô∏è Outliers Present in the dataset
Used Boxplot to visualize and detect outliers for numerical features like Age, Income, Miles & Usage

```python
# BoxPlot for Age
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Age"])
plt.title("Boxplot of Age")
plt.show()
```

#### üîç Insights
* Most customers who purchase treadmills fall between 22 and 40 years old.
* A small group of older customers (ages 47‚Äì50) might represent fitness enthusiasts or individuals who prefer high-quality treadmills like KP781.

```python
# Boxplot for Income
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Income"])
plt.title("Boxplot of Income")
plt.show()
```

### üîç Insights
* Most customers have incomes between 40,000 and 70,000, forming the core target audience.
* Customers with incomes above 75,000 are outliers, but they represent a niche, high-income group.
* This group of high-income customers might be more likely to purchase premium products like the KP781 treadmill, which has advanced features and sells for 2,500.

```python
# Boxplot for Miles
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Miles"])
plt.title("Boxplot of Miles")
plt.show()
```

#### üîç Insights
* Most users log between 50 and 150 miles, forming the primary usage range.
* This segment represents consistent treadmill usage by regular customers.
* The outliers with values above 180 miles highlight a niche group of extreme users.
* These individuals may have higher fitness goals or performance demands, such as athletes or fitness enthusiasts.

```python
# Boxplot for Usage
plt.figure(figsize=(8, 5))
sns.boxplot(x=df["Usage"])
plt.title("Boxplot of Usage")
plt.show()
```

#### üîç Insights
* Most customers use the treadmill between 2 and 5 days per week, forming the primary usage group.
* This segment likely includes moderate users who seek consistent but not extreme fitness routines.
* Outliers with usage values of 6 and 7 days per week represent highly dedicated users.

```python
numerical_df = df.select_dtypes(include=["number"])
plt.figure(figsize=(10, 6))
sns.heatmap(numerical_df.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

#### üîç Insights
1) Strong Positive Correlations:
   * Income vs Fitness: A strong positive correlation close to 1 indicates that individuals with higher fitness levels also tend to have higher incomes.
   * Miles vs Fitness: Miles and Fitness have a high positive correlation. This suggests that customers who walk or run more miles also have higher fitness levels.
   * Income vs Miles: Income and Miles also have a positive correlation, meaning people with higher incomes tend to cover more miles.
2) Weak Correlations:
   * Age vs Miles: There is a very weak correlation (~0.037) between Age and Miles, suggesting age does not strongly influence the number of miles walked.
   * Age vs Usage: Similarly, Age and Usage have a near-zero correlation (~0.015), indicating that age has no significant relationship with the frequency of product usage.
3) Moderate Correlation:
   * Age vs Income: A moderate correlation (~0.51) suggests that older individuals might have higher incomes, which could be attributed to career progression over time.
4) Education:
   * Education has moderate correlations with Income and Fitness, indicating that higher education levels might have some influence on fitness and income but not very strongly.

Based on your analysis of the dataset, I have created detailed customer profiles for each treadmill product: KP281, KP481, and KP781.

#### 1) KP281: Entry-Level Treadmill

  Younger individuals with lower income, moderate fitness levels, and limited weekly usage.

     * Weekly mileage is lower, suggesting limited use, likely for entry-level workouts.

#### 2) KP481: Mid-Range Treadmill
* Target Customers: Middle-aged individuals with moderate-to-high fitness levels and moderate income.

  2) Income:
     * Customers have moderate income levels.

  3) Fitness and Usage:
     * KP481 buyers show higher fitness levels and usage compared to KP281 customers.

#### 3) KP781: Advanced Treadmill
